```sql
SELECT
    region,
    ROUND(total_sales,2) AS total_sales,
    ROUND(total_profit,2) AS total_profit
FROM retail_lakehouse.gold.sales_by_region
ORDER BY total_sales DESC;
```

```sql
SELECT
    category,
    ROUND(total_sales,2) AS total_sales,
    ROUND(total_profit,2) AS total_profit
FROM retail_lakehouse.gold.sales_by_category
ORDER BY total_sales DESC;
```

```sql
SELECT
    segment,
    ROUND(total_sales,2) AS total_sales,
    ROUND(total_profit,2) AS total_profit
FROM retail_lakehouse.gold.sales_by_segment
ORDER BY total_sales DESC;
```

```sql
SELECT
    product_name,
    ROUND(total_sales,2) AS total_sales
FROM retail_lakehouse.gold.top_products
ORDER BY total_sales DESC;
```

```sql
SELECT
    discount_num,
    ROUND(total_sales,2) AS total_sales,
    ROUND(total_profit,2) AS total_profit
FROM retail_lakehouse.gold.discount_analysis
ORDER BY discount_num;
```

```sql
SELECT
    state,
    ROUND(total_profit,2) AS total_profit
FROM retail_lakehouse.gold.state_profit
ORDER BY total_profit DESC;
```

```sql
SELECT
    product_name,
    ROUND(total_profit,2) AS total_profit
FROM retail_lakehouse.gold.profitable_products
ORDER BY total_profit DESC;
```

```sql
SELECT
    product_name,
    ROUND(total_profit,2) AS total_profit
FROM retail_lakehouse.gold.loss_products
ORDER BY total_profit;
```

```sql
SELECT
    customer_name,
    ROUND(total_sales,2) AS total_sales
FROM retail_lakehouse.gold.customer_sales
ORDER BY total_sales DESC
LIMIT 15;
```

```sql
SELECT
    category,
    ROUND(
        total_sales * 100 /
        SUM(total_sales) OVER (),
        2
    ) AS sales_percentage
FROM retail_lakehouse.gold.sales_by_category
ORDER BY sales_percentage DESC;
```
