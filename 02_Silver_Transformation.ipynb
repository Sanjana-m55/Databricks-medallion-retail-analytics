```sql
%python
#%python

bronze_df = spark.table(
    "retail_lakehouse.bronze.superstore_raw"
)
```

```sql
%python
#%python

print(
    "Bronze Record Count:",
    bronze_df.count()
)
```

```sql
%python
#%python

from pyspark.sql.functions import *

silver_df = (
    bronze_df

    # Remove duplicate rows
    .dropDuplicates()

    # Standardize text columns
    .withColumn("customer_name", trim(col("customer_name")))
    .withColumn("city", trim(col("city")))
    .withColumn("state", trim(col("state")))
    .withColumn("category", trim(col("category")))
    .withColumn("sub_category", trim(col("sub-category")))

    # Create business metric
    .withColumn(
        "shipping_days",
        datediff(col("ship_date"), col("order_date"))
    )
)
```

```sql
%python
#%python

silver_df.printSchema()
```

```sql
%python
#%python

print(
    "Silver Record Count:",
    silver_df.count()
)
```

```sql

DROP TABLE IF EXISTS retail_lakehouse.silver.superstore_clean;
```

```sql
%python
#%python

silver_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "retail_lakehouse.silver.superstore_clean"
    )
```

```sql
%python
#%python

display(
    spark.table(
        "retail_lakehouse.silver.superstore_clean"
    )
)
```

```sql
%python
#%python

print("Silver Layer Created Successfully")
print("Data Quality Improvements Applied:")
print("- Duplicate check performed")
print("- Text standardization applied")
print("- Shipping Days feature created")
print("- Silver table stored in Delta format")
```
