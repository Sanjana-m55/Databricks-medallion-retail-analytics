#Dashboard Visualizations

This dashboard is built on Gold Layer business tables generated using Medallion Architecture.

# Visualization 1: Sales by Region

**Chart Type:** Bar Chart

**Source Table:**
retail_lakehouse.gold.sales_by_region

**Purpose:**
Compare total sales and profit across different regions.

This visualization compares total sales across different regions of the United States.

It helps identify which geographical region contributes the most revenue to the business. Understanding regional sales performance enables organizations to focus marketing efforts, inventory planning, and resource allocation on high-performing markets.

```sql
SELECT
    region,
    total_sales
FROM retail_lakehouse.gold.sales_by_region
ORDER BY total_sales DESC;
```

#Visualization 4: Top 10 Products by Sales

Chart Type: Horizontal Bar Chart

Source Table: retail_lakehouse.gold.top_products

Purpose:
Identify the highest revenue-generating products in the business.

Business Insight:
The analysis highlights the top 10 products contributing the highest sales revenue. These products represent the strongest-performing items in the portfolio and can be prioritized for inventory planning, marketing campaigns, and sales strategies.

```sql
SELECT
    product_name,
    total_sales
FROM retail_lakehouse.gold.top_products
ORDER BY total_sales DESC;
```
