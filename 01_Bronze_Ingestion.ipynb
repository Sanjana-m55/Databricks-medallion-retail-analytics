```sql
%python

df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")
         .csv("/Volumes/retail_lakehouse/default/raw_data/Sample - Superstore 2.csv")
)

display(df)
```

```sql
%python

df_clean = df.toDF(
    *[
        c.lower().replace(" ", "_")
        for c in df.columns
    ]
)

display(df_clean)
```

```sql
%python

df_clean.printSchema()
```

```sql
%python

print(
    "Raw Record Count:",
    df_clean.count()
)
```

```sql
%python

df_clean.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "retail_lakehouse.bronze.superstore_raw"
    )
```

```sql
%python

display(
    spark.table(
        "retail_lakehouse.bronze.superstore_raw"
    )
)
```

```sql
%python

print(
    "Bronze Table Rows:",
    spark.table(
        "retail_lakehouse.bronze.superstore_raw"
    ).count()
)
```
